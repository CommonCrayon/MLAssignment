```python
# Assuming the dataset is in the same directory as your notebook
df = pd.read_csv('dataset.csv')
```

```python
# Count occurrences for all unique values in the column
print(df['Target'].value_counts())
```

```python
# Get 1400 of graduate and droput:
graduate_df = df[df['Target'] == 'Graduate'].head(1400)
dropout_df = df[df['Target'] == 'Dropout'].head(1400)

# Combine the dataframes
combined_df = pd.concat([graduate_df, dropout_df])

# Reset the index to ensure that the rows are properly randomized
combined_df = combined_df.reset_index(drop=True)

# Shuffle (randomize) the rows
combined_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Now, combined_df contains the combined and randomized data
```

```python
X = df.drop(columns=['Target', 'Curricular units 1st sem (enrolled)', 'Curricular units 2nd sem (enrolled)']).to_numpy()
y = df['Target']
```

```python
X_reshaped = X.reshape(-1, 8, 4)
print(X_reshaped.shape)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_final = to_categorical(y_encoded, num_classes=3)
print(y_final)
```

```python
from sklearn.model_selection import train_test_split

# Splitting a dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_final, test_size=0.2, random_state=42)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Flatten(input_shape=(8, 4)),  # Flatten the 8x4 input into a 32-element vector
    layers.Dense(128, activation='relu'),  # 64 neurons in the hidden layer with ReLU activation
    layers.Dense(128, activation='relu'),  # 64 neurons in the hidden layer with ReLU activation
    layers.Dense(128, activation='relu'),  # 64 neurons in the hidden layer with ReLU activation
    layers.Dense(3, activation='softmax')  # Output layer with softmax activation for classification
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
